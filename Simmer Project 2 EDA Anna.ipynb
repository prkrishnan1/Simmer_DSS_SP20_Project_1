```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

# Import Data

```python
AugSep19 = pd.read_csv('Simmer_AugSep19.csv')
DecJan19 = pd.read_csv('Simmer_DecJan19.csv')
FebMar20 = pd.read_csv('Simmer_FebMar20.csv')
OctNov19 = pd.read_csv('Simmer_OctNov19.csv')
frequent_users = pd.read_csv('People_Sample.csv') #more than thirty events
```

```python
AugSep19Size = AugSep19.shape
DecJan19Size = DecJan19.shape
FebMar20Size = FebMar20.shape
OctNov19Size = OctNov19.shape
```

```python
#append monthly data into main data frame
simmer_data_app = AugSep19.append(OctNov19)
simmer_data_app = simmer_data_app.append(DecJan19)
simmer_data_app = simmer_data_app.append(FebMar20)
```

```python
AugSep19Size[0]+DecJan19Size[0]+FebMar20Size[0]+OctNov19Size[0] == simmer_data_app.shape[0] #check size is correct
```

# Start of Cleaning and EDA

```python
threshold = simmer_data_app.shape[0] * .95 # keep if less than 95 percent of rows are na
ninetyfive_boolean = simmer_data_app.isna().sum()<threshold
ninetyfive_boolean = ninetyfive_boolean.values
```

```python
simmer_data_app_cleaned = simmer_data_app.loc[:, ninetyfive_boolean]
simmer_data_app_cleaned.groupby('properties.mp_country_code')
```

```python
#manually dropping columns that dont seem useful to churn. Do more research on this
simmer_data_app_cleaned = simmer_data_app_cleaned.drop(['dataset', 'properties.$app_build_number','properties.$app_release','properties.$app_version','properties.$app_version_string',
                                                       'properties.$os','properties.$os_version','properties.mp_device_model',
                                                        'properties.$screen_width','properties.$screen_dpi','properties.$screen_height',
                                                       'sampling_factor','properties.$carrier', 'properties.$bluetooth_version','properties.$brand',
                                                       'properties.$google_play_services', 'properties.$ios_ifa','properties.$distinct_id_before_identity',
                                                       ],axis=1)
simmer_data_app_cleaned
```
