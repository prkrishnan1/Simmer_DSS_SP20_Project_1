```python
import numpy as np
import pandas as pd
```

```python
aug_sep = pd.read_csv("Simmer_AugSep19.csv", low_memory=False)
oct_nov = pd.read_csv("Simmer_OctNov19.csv", low_memory=False)
dec_jan = pd.read_csv("Simmer_DecJan19.csv", low_memory=False)
feb_mar = pd.read_csv("Simmer_FebMar20.csv", low_memory=False)
people = pd.read_csv("People_Sample.csv", low_memory=False)
```

```python
def filter_nan(df, threshold):
    length = len(df)
    result = pd.DataFrame()
    for col in df.columns:
        if (df[col].isnull().sum() < length * threshold):
            result[col] = df[col]
    return result
```

```python
filtered_aug_sep = filter_nan(aug_sep, 0.9)
filtered_oct_nov = filter_nan(oct_nov, 0.9)
filtered_dec_jan = filter_nan(dec_jan, 0.9)
filtered_feb_mar = filter_nan(feb_mar, 0.9)
```

```python
print(filtered_aug_sep.columns.size, filtered_oct_nov.columns.size, filtered_dec_jan.columns.size, filtered_feb_mar.columns.size)
```

```python
total = filtered_aug_sep.append(filtered_oct_nov).append(filtered_dec_jan).append(filtered_feb_mar)
len(total)
```
