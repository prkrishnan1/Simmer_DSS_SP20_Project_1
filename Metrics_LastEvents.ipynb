```python
events = pd.read_csv('Cleaned_Dataset_With_ID.csv', low_memory=False)
people = pd.read_csv('labeled_data.csv', low_memory= False)
```

```python
people = people.drop(columns=['Unnamed: 0'])[['distinct_id', 'churned']]
people
```

```python
df = pd.merge(events, people, left_on = 'distinct_id', right_on = 'distinct_id', how = 'left').drop(columns=['Unnamed: 0'])
df 
```

```python
grouped = df.groupby(['distinct_id', 'session_id'], as_index=False).agg({'name_x' : list, 'churned' : 'last'})
#group_user_session = group_user_session.rename(columns = {'name_x' : 'event_count', 'score' : 'SQS'})
```

```python
grouped
```

```python
last_id = grouped.groupby(['distinct_id'], as_index=False).agg({'session_id' : max, 'name_x' : 'last', 'churned' : 'last'})
```

```python
last_id['length'] = last_id['name_x'].apply(lambda x: len(x))
last_id
```

```python
churned = last_id[last_id['churned'] == 1]
churned.head()
```
